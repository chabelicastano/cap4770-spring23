```python
#import pandas
import pandas as pd
```

```python
# I saved the csv file in colab
# read 'bach.csv'
bach_df = pd.read_csv('bach.csv')
bach_df
```

```python
#check the labels
bach_df['chord_label'].unique()
```

```python
# check for null values
bach_df.isnull().values.any()
```

```python
# Encode Binary Categorical columns (notes)
binary_cat_columns = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
for col in binary_cat_columns:
  bach_df[col] = bach_df[col].replace({'YES': 1, 'NO': 0})

bach_df
```

```python
# Since some chord_labels occur only once, we are going to drop it to avoid them being 
# only in the test and not in the train, and viceversa

#count the ocurrences of each cord_label and add it as a column to the dataframe
bach_df['chord_label_count'] = bach_df.groupby('chord_label')['chord_label'].transform('count')
bach_df
```

```python
# I need to count how many rows will be dropped, to make sure afterwards
# that it was properly done

#display the entries whose chord_label_count is 1
bach_df.loc[(bach_df['chord_label_count']== 1)]
```

So, we have 8 rows where the label count is 1. After we dropped then, we should have 5657 rows in our bach_df

```python
#remove the entries where chord_label_count == 1
bach_df = bach_df[bach_df.chord_label_count != 1]
bach_df
```

As we can see 8 entries were succesfully removed and now we have 5657 rows in the bach_df

```python
# Now, let's encode 'bass' since this is a categorical column

#import OneHotEncoding from sklearn
from sklearn.preprocessing import OneHotEncoder

#create an encoder instance
ohe = OneHotEncoder(handle_unknown='ignore')
```

```python
#create a df that only contains the 'bass' column, to make data handling easier
bass_df = bach_df[['bass']]
bass_df
```

```python
# transform the column
transformed = ohe.fit_transform(bass_df[['bass']])

#create a dataframe with the columns resulting from the OneHotEncoding transformation
bass_2 = pd.DataFrame.sparse.from_spmatrix(transformed, columns=ohe.get_feature_names_out())
bass_2
```

```python
# join the onehotencoded columns to the original dataframe

# name of the columns in the bass_2 df
bass2_cols = list(bass_2.columns)

bass2_cols
```

```python
# join those transformed columns to the bach_df
for col in bass2_cols:
  bach_df[col] = bass_2[col].values
bach_df
```

```python
#check for null values
# I'm doing this because for some of my tries joining the columns I got some NaN values
# out of nowhere, so I want to make sure that the copy was succesful
bach_df.isnull().values.any()
```

```python
# Let's drop the original 'bass' column
bach_df = bach_df.drop('bass', axis=1)
bach_df
```

```python
# let's label encode the chord_label column

#import LabelEncoder
from sklearn.preprocessing import LabelEncoder

#create a label encoder object
le = LabelEncoder()

#do the encoding
bach_df['chord_label'] = le.fit_transform(bach_df['chord_label'])
bach_df['chord_label']
```

```python
#divide train and test

#import for train_test_split
from sklearn.model_selection import train_test_split

bach_train, bach_test = train_test_split(bach_df, test_size = 0.2, random_state=42, stratify= bach_df['chord_label'])
```

```python
# features and labels

# get the names of the columns
bach_columns = list(bach_df.columns)
bach_columns
```

```python
# for features, I will use all the columns except for: 'choral_id', 'event_number', 'chord_label' 
# (which is the label), and 'chord_label_count'
# So, let's remove those from the list of column names
bach_columns.remove('choral_ID')
bach_columns.remove('event_number')
bach_columns.remove('chord_label')
bach_columns.remove('chord_label_count')

bach_columns
```

```python
#features
#train
bach_features_train = bach_train[bach_columns]
#test
bach_features_test = bach_test[bach_columns]
```

```python
#labels
#train
bach_labels_train = bach_train['chord_label']
#test
bach_labels_test = bach_test['chord_label']
```

```python
#imports
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
```

```python
# XGBoost classifier
from xgboost import XGBClassifier
xgb = XGBClassifier()
```

```python
#import for numpy - necessary for the arrays in the param_grid
import numpy as np
```

```python
#param_grid
# I will be tuning the following parameters: 'learning_rate', 'max_depth', 'n_estimators'

param_grid = {
    'learning_rate': [0.0001,0.001, 0.01, 0.1, 1],
    'max_depth': np.arange(3, 21, 2),
    'n_estimators': np.arange(50, 400, 50),
    'tree_method': ['gpu_hist'], 
    'predictor' : ['gpu_predictor']
}
```

```python
#import stratified kfold
from sklearn.model_selection import StratifiedKFold

#create the object
k_fold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
#define gridsearch
random_search = RandomizedSearchCV(estimator=xgb,
                           param_distributions=param_grid,
                           n_iter = 15,
                           cv=k_fold,
                           n_jobs=-1,
                           verbose=3)
```

```python
#check gpu
!nvidia-smi
```

```python
#fit the random_search
random_search.fit(bach_features_train, bach_labels_train)
```

```python
#Let's check which were the best parameters
random_search.best_params_
```

```python
#Let's predict
bach_predictions = random_search.best_estimator_.predict(bach_features_test)
```

```python
#import for the accuracy
from sklearn.metrics import accuracy_score
#Finally, let's calculate the accuracy of the results
accuracy = accuracy_score(bach_labels_test, bach_predictions)
print('The accuracy score is: {:.4f}'.format(accuracy))
```
